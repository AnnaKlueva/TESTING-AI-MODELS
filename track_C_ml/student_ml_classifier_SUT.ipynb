Модель `sut_artifact.joblib` і мічений набір `production_data.csv` **уже лежать у цій теці** (в репозиторії),
нічого завантажувати не треба. Працюй через `predict` / `predict_proba`.

```python
import joblib
import pandas as pd

class SystemUnderTest:
    def __init__(self, path='sut_artifact.joblib'):
        bundle = joblib.load(path)
        self._pipe = bundle['pipeline']
        self.features = bundle['features']
    def predict(self, df):
        return self._pipe.predict(df[self.features])
    def predict_proba(self, df):
        return self._pipe.predict_proba(df[self.features])[:, 1]

sut = SystemUnderTest()
prod = pd.read_csv('production_data.csv')
print(prod.shape, list(prod.columns))
```

```python
sample = prod.head(3)
print('preds:', sut.predict(sample))
print('proba:', sut.predict_proba(sample).round(3))
```
